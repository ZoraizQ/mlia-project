```python
import SimpleITK as sitk
import matplotlib.pylab as plt
import os


def visualize(image, label):
    plt.figure(figsize=(8,8))
    plt.imshow(image, cmap='gray')

    alpha = label.copy()
    alpha[alpha>0] = 1
    plt.imshow(label, alpha=alpha, cmap='jet')
    plt.axis('off')
    plt.show()


TRAIN_IMAGES = 'Training/Brains'
TRAIN_LABELS = 'Training/Labels'
TEST_IMAGES = 'Testing/Brains'
TEST_LABELS = 'Testing/Labels'

train_ids = [int(fpath.split('_')[1].split('.')[0]) for fpath in os.listdir(TRAIN_IMAGES) if fpath.endswith('.raw')]
train_ids.sort()

test_ids = [int(fpath.split('_')[1].split('.')[0]) for fpath in os.listdir(TEST_IMAGES) if fpath.endswith('.raw')]
test_ids.sort()

print(len(train_ids), 'train', len(test_ids), 'test') 
print('file_id examples:', train_ids[:30])
```

```python
id = 0 # set this to the image/subject ID

image = sitk.GetArrayFromImage(sitk.ReadImage(os.path.join(TRAIN_IMAGES, f"mri_{id}.mhd"), sitk.sitkFloat32))
label = sitk.GetArrayFromImage(sitk.ReadImage(os.path.join(TRAIN_LABELS, f"seg_{id}.mhd"), sitk.sitkFloat32))

visualize(image, label)
```
